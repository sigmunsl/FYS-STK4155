```python
# create dataset and create design matrix
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as skl
import sklearn.metrics as sklm


x = np.expand_dims(np.sort(np.random.rand(100,1), None), -1)
y = 5*x*x+0.1*np.random.randn(100,1)

X = np.empty((len(x), 3))

X[:, 0] = 1
X[:, 1] = x[:, 0]
X[:, 2] = x[:, 0]**2
```

```python
# define lambda and invert

la             = 0.1
toinvert       = X.T.dot(X) + la*np.eye((X.shape[1]))
beta_ridge     = np.linalg.inv(toinvert).dot(X.T).dot(y)
beta_ols       = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

ytilde_ridge   = X @ beta_ridge
ytilde_ols     = X @ beta_ols

```

```python
plt.scatter(x, y)
plt.plot(x, ytilde_ols, color='red', label='OLS')
plt.plot(x, ytilde_ridge, color='green', label='Ridge')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 4 B

```python

# repeating the same regressions, but utilizing the scikit-learn package

beta_skl_OLS      = skl.LinearRegression().fit(X, y)
beta_skl_Ridge    = skl.Ridge(alpha=0.1).fit(X, y)
ytilde_skl_OLS    = beta_skl_OLS.predict(X)
ytilde_skl_Ridge  = beta_skl_Ridge.predict(X)

plt.scatter(x, y)
plt.plot(x, ytilde_skl_OLS, color='red', label='OLS')
plt.plot(x, ytilde_skl_Ridge, color='green', label='Ridge')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 4 C

```python
# we will now find the variance of parameters beta


def var_beta(X, _lambda):
    toinvert       = X.T.dot(X) + _lambda*np.eye((X.shape[1]))
    variance       = np.linalg.inv(toinvert)# ikke egentlig varianse (dette er bare en matrise)
    return np.diag(variance)
```

```python
_lambdavec = np.linspace(0,2,50)
vb         = np.empty((len(_lambdavec),3))
for i in range(len(_lambdavec)):
    vb[i] = var_beta(X,_lambdavec[i])
```

```python
plt.plot(_lambdavec,vb[:,1], label=r'$\beta_1$')
plt.plot(_lambdavec,vb[:,2], label=r'$\beta_2$')
plt.legend()
plt.xlabel(r'$\lambda$')
plt.ylabel('Variance')
plt.show()
```

## 4 E

```python
# compute mse and R2 score


def mse_lasso_ridge(_lambda,X,y):
    lass       = skl.Lasso(alpha=_lambda).fit(X,y)
    lass_pred  = lass.predict(X)
    lass_mse   = sklm.mean_squared_error(y,lass_pred)
    
    ridge      = skl.Ridge(alpha=_lambda).fit(X,y)
    ridge_pred = ridge.predict(X)
    ridge_mse  = sklm.mean_squared_error(y,ridge_pred)
    return lass_mse, ridge_mse
```

```python
mserrors = np.empty((len(_lambdavec),2))
for i in range(len(_lambdavec)):
    mserrors[i]= mse_lasso_ridge(_lambdavec[i], X,y)
```

```python
plt.plot(_lambdavec,mserrors[:,0], label='Lasso')
plt.plot(_lambdavec,mserrors[:,1], label='Ridge')
plt.xlabel(r'$\lambda$')
plt.ylabel('MSE')
plt.show()
```
