```python
# create dataset and create design matrix
import numpy as np
import matplotlib.pyplot as plt
x = np.expand_dims(np.sort(np.random.rand(100,1), None), -1)
y = 5*x*x+0.1*np.random.randn(100,1)

X = np.empty((len(x), 3))

X[:, 0] = 1
X[:, 1] = x[:, 0]
X[:, 2] = x[:, 0]**2
```

```python
# define lambda and invert

la       = 0.1
toinvert = X.T.dot(X) + la*np.eye((X.shape[1]))
beta     = np.linalg.inv(toinvert).dot(X.T).dot(y)

ytilde   = X @ beta

```

```python
plt.scatter(x, y)
plt.plot(x, ytilde, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
